```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# %% create dataset

# class1
x1 = np.random.normal(25,5,10000)
y1 = np.random.normal(25,5,10000)

# class2
x2 = np.random.normal(55,5,10000)
y2 = np.random.normal(60,5,10000)

# class3
x3 = np.random.normal(55,5,10000)
y3 = np.random.normal(15,5,10000)
```

```python
x = np.concatenate((x1,x2,x3),axis = 0)
y = np.concatenate((y1,y2,y3),axis = 0)
```

```python
dictionary = {"x":x,"y":y}

data = pd.DataFrame(dictionary)

plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
plt.show()

## %% kmeans algoritmasÄ± bunu gorecek
#plt.scatter(x1,y1,color = "black")
#plt.scatter(x2,y2,color = "black")
#plt.scatter(x3,y3,color = "black")
#plt.show()
```

```python
# %% KMEANS

from sklearn.cluster import KMeans
wcss = []

for k in range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,15),wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Veri Ã¶rneÄi oluÅturma
np.random.seed(0)
data = np.random.rand(100, 2)

# WCSS deÄerlerini hesaplama
wcss = []

for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# WCSS deÄerlerini grafiÄe Ã§izme
plt.figure(figsize=(8, 6))
plt.plot(range(1, 15), wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method")
plt.grid(True)
plt.show()
```

```python
#%% k = 3 icin modelim

kmeans2 = KMeans(n_clusters=3)
clusters = kmeans2.fit_predict(data)

data["label"] = clusters

plt.scatter(data.x[data.label == 0 ],data.y[data.label == 0],color = "red")
plt.scatter(data.x[data.label == 1 ],data.y[data.label == 1],color = "green")
plt.scatter(data.x[data.label == 2 ],data.y[data.label == 2],color = "blue")
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color = "yellow")
plt.show()
```

```python
# Veri Ã¶rneÄi oluÅtur
np.random.seed(0)
data = np.random.rand(1000, 2)

# K-Means modelini oluÅtur ve verileri kÃ¼melere ayÄ±r
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(data)

# Verileri bir DataFrame'e eklemek yerine numpy dizileri olarak kullanÄ±yoruz
# Bu nedenle, aÅaÄÄ±daki iÅlemde data DataFrame'i kullanmayacaÄÄ±z
# data["label"] = clusters

# KÃ¼me merkezlerini al
cluster_centers = kmeans.cluster_centers_

# KÃ¼me merkezlerini ve verileri grafiÄe Ã§iz
plt.scatter(data[clusters == 0, 0], data[clusters == 0, 1], color='red', label='Cluster 1')
plt.scatter(data[clusters == 1, 0], data[clusters == 1, 1], color='green', label='Cluster 2')
plt.scatter(data[clusters == 2, 0], data[clusters == 2, 1], color='blue', label='Cluster 3')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='yellow', marker='o', s=150, label='Cluster Centers')
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("K-Means Clustering (k=3)")
plt.grid(True)
plt.show()
```
